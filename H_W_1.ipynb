## Задание № 1

Мной выбран продукт - Ozon.ru

Метрика № 1. 
GMV (Gross Merchandise Volume) - сумма всех товарных объемов, проданных на платформе Ozon.ru за определенный период. Это включает в себя сумму всех заказов, включая товары и услуги.

Метрика № 2. 
Количество заказов - общее количество заказов, совершенных покупателями на платформе.

Метрика № 3. 
Средний чек - средняя стоимость каждого заказа на платформе. Рассчитывается как общая сумма выручки, деленная на количество заказов.

Метрика № 4. 
Конверсия - процент пользователей, посещающих сайт, оформляющих покупку или выполняющих желаемое действие. Рассчитывается как отношение числа совершенных заказов к общему числу посещений сайта.

Метрика № 5. 
Коэффициент возвратов - процент товаров или услуг, возвращенных покупателями из-за неудовлетворительного качества, неподходящего размера или других причин.

Метрика № 6. 
Количество активных пользователей - количество зарегистрированных пользователей, которые посещают платформу и выполняют действия, такие как оформление заказов, покупки или оставление отзывов.

Метрика № 7. 
Время нахождения на сайте - среднее время, которое пользователи проводят на сайте или в приложении ozon.ru. Эта метрика может указывать на уровень интереса пользователей к контенту и функциональности платформы.

Метрика № 8. 
Уровень удовлетворенности клиентов - средняя оценка или отзывы, оставленные покупателями после совершения покупки или получения услуги на платформе ozon.ru. Это может включать рейтинг продавцов, привлекательность интерфейса пользователя или качество услуг.

Метрика № 9. 
Коэффициент повторных покупок - процент покупателей, совершающих повторные покупки на платформе ozon.ru. Это может свидетельствовать о лояльности пользователей и уровне удовлетворенности.

Метрика № 10. 
Уровень участия продавцов - количество и качество продавцов, предоставляющих свои товары и услуги на платформе Ozon.ru. Включает в себя рейтинги продавцов, объем продаж и уровень удовлетворенности клиентов.

## Задание № 2

Гипотеза № 1. 
Если упростить процесс оформления заказа и сократить количество шагов, то уровень завершенности заказа повысится.

Гипотеза № 2. 
Если предоставить подробную информацию о продукте, включая отзывы, характеристики и фотографии, то повысится уровень доверия пользователей и вероятность совершения покупки.

Гипотеза № 3. 
Если предложить программу лояльности для постоянных клиентов с бонусами и скидками, то уровень повторных покупок увеличится.

Гипотеза № 4. 
Если улучшить пользовательский интерфейс и навигацию на сайте, то снизится количество отказов и увеличится конверсия.

Гипотеза № 5. 
Если добавить больше фильтров и возможностей сортировки при поиске товара, то пользователи смогут быстрее найти желаемый товар, что повысит удовлетворенность и вероятность покупки.

Гипотеза № 6. 
Если улучшить систему рекомендаций товаров на основе предпочтений пользователя и истории покупок, то повысится вероятность дополнительных продаж.

Гипотеза № 7. 
Если предложить различные способы оплаты, включая электронные кошельки и рассрочку, то расширится число пользователей, способных совершить покупку.

Гипотеза № 8. 
Если улучшить процесс доставки, предоставить бесплатную доставку или сократить время доставки, то повысится удовлетворенность клиентов и лояльность к платформе.

Гипотеза № 9. 
Если разработать мобильное приложение с дополнительными функциями, такими как push-уведомления о скидках и акциях, то повысится вовлеченность пользователей и их активность.

Гипотеза № 10. 
Если предоставить возможность возвращать товары без проблем, с упрощенным процессом возврата и возврата денежных средств, то повысится доверие пользователей и уровень повторных покупок.
